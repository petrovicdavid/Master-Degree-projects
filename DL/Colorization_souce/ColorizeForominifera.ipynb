```python
import os
import sys
import cv2
import numpy as np


mother_folder_path = r"data\Foraminifera"  # Replace with the actual folder path
cout = 0
# Get the list of all folders in the specified path
species_folder_names = [folder for folder in os.listdir(mother_folder_path) if os.path.isdir(os.path.join(mother_folder_path, folder))]
for species_folder_name in species_folder_names:
    
    species_folder_path = os.path.join(mother_folder_path, species_folder_name)

    # Get the list of all folders in the specified path
    folder_names = [folder for folder in os.listdir(species_folder_path) if os.path.isdir(os.path.join(species_folder_path, folder))]

    for folder_name in folder_names:
        if folder_name == "mean_images":
            break
        folder_path = os.path.join(species_folder_path, folder_name)
        image_files = os.listdir(folder_path)

        # Initialize an empty list to store the images
        images = []

        # Loop through the image files
        for image_file in image_files:
            # Construct the full path to the image file
            image_path = os.path.join(folder_path, image_file)

            # Read the image
            image = cv2.imread(image_path)

            # Check if image reading was successful
            if image is not None:
                # Append the image to the list
                images.append(image)

        # Check if any images were successfully read
        if len(images) == 0:
            print("No valid images found in the folder.")
            exit()

        # Calculate the mean image
        mean_image = images[0]
        height, width, channels = mean_image.shape
        if len(images) > 0:
            # Get the dimensions of the first image
            height, width, channels = images[0].shape

            # Create an array to store the mean image
            mean_image = np.zeros((height, width, channels), dtype=np.float32)

            # Loop over the rows and columns
            """ for v in range(height):
                for u in range(width):
                    
                    pixel_sum = np.sum([image[v, u] for image in images], axis=0)
                    mean_image[v, u] = pixel_sum / len(images)"""
            d=2
            for v in range(height):
                for u in range(width):
                    pixels = [image[v, u] for image in images]
                    sorted_pixels = np.sort(pixels)
                    trimmed_pixels = sorted_pixels[int(d/2):len(sorted_pixels)-int(d/2)]  # Remove the smallest and largest 'd' pixels
                    pixel_sum = np.sum(trimmed_pixels, axis=0)
                    mean_image[v, u] = pixel_sum / len(trimmed_pixels)

        mean_image = mean_image.astype(np.uint8)

        # Save the image as jpg
        mean_path = os.path.join(species_folder_path, "mean_images")
        if not os.path.exists(mean_path):
            os.makedirs(mean_path)
        mean_path = os.path.join(mean_path, folder_name + ".jpg")
        cv2.imwrite(mean_path, mean_image)
        cout += 1
        print("done image n",cout)
        
        
```

```python
"""
Credits: 
	1. https://github.com/opencv/opencv/blob/master/samples/dnn/colorization.py
	2. http://richzhang.github.io/colorization/
	3. https://github.com/richzhang/colorization/
"""

"""
Download the model files: 
	1. colorization_deploy_v2.prototxt:    https://github.com/richzhang/colorization/tree/caffe/colorization/models
	2. pts_in_hull.npy:					   https://github.com/richzhang/colorization/blob/caffe/colorization/resources/pts_in_hull.npy
	3. colorization_release_v2.caffemodel: https://www.dropbox.com/s/dx0qvhhp5hbcx7z/colorization_release_v2.caffemodel?dl=1

"""
import os
import time
import cv2
import numpy as np

cont=0

# Paths to load the model
DIR = r"C:\Users\usalv\Documents"
PROTOTXT = os.path.join(DIR, r"model/colorization_deploy_v2.prototxt")
POINTS = os.path.join(DIR, r"model/pts_in_hull.npy")
MODEL = os.path.join(DIR, r"model/colorization_release_v2.caffemodel")


# Load the Model
print("Load model")
net = cv2.dnn.readNetFromCaffe(PROTOTXT, MODEL)
pts = np.load(POINTS)

# Load centers for ab channel quantization used for rebalancing.
class8 = net.getLayerId("class8_ab")
conv8 = net.getLayerId("conv8_313_rh")
pts = pts.transpose().reshape(2, 313, 1, 1)
net.getLayer(class8).blobs = [pts.astype("float32")]
net.getLayer(conv8).blobs = [np.full([1, 313], 2.606, dtype="float32")]

mother_folder_path = r"data\Foraminifera"
son_folder_path = r"data\ForaminiferaColored"
if not os.path.exists(son_folder_path):
            os.makedirs(son_folder_path)
species_folder_names = [folder for folder in os.listdir(mother_folder_path) if os.path.isdir(os.path.join(mother_folder_path, folder))]
for species_folder_name in species_folder_names:
    #start timer	
	species_folder_path = os.path.join(mother_folder_path, species_folder_name)
	son_species_folder_path = os.path.join(son_folder_path, species_folder_name)
 
	#for every mean image in the meanfolder 
	mean_image_folder_path = os.path.join(species_folder_path, "mean_images")
	mean_image_files = os.listdir(mean_image_folder_path)
	# #for every mean image in the meanfolder
	for mean_image_file in mean_image_files:
	# Load the input image
		image_path = os.path.join(mean_image_folder_path, mean_image_file)
		image = cv2.imread(image_path)
		scaled = image.astype("float32") / 255.0
		lab = cv2.cvtColor(scaled, cv2.COLOR_BGR2LAB)

		resized = cv2.resize(lab, (224, 224))
		L = cv2.split(resized)[0]
		L -= 50

		net.setInput(cv2.dnn.blobFromImage(L))
		ab = net.forward()[0, :, :, :].transpose((1, 2, 0))

		ab = cv2.resize(ab, (image.shape[1], image.shape[0]))

		L = cv2.split(lab)[0]
		colorized = np.concatenate((L[:, :, np.newaxis], ab), axis=2)

		colorized = cv2.cvtColor(colorized, cv2.COLOR_LAB2BGR)
		colorized = np.clip(colorized, 0, 1)

		colorized = (255 * colorized).astype("uint8")
		# Save the image as jpg
		if not os.path.exists(son_species_folder_path):
			os.makedirs(son_species_folder_path)
		mean_image_file=mean_image_file.replace(".jpg",".png")
		colorized_path = os.path.join(son_species_folder_path, mean_image_file)
		cv2.imwrite(colorized_path, colorized)
  
		cont+=1
		
		print("Done cont: ",cont)
		
  
```
